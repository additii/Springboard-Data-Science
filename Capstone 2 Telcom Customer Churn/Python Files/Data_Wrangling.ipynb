## 2.6 Data Profiling<a id='2.6_Data_profiling'></a>

```python
#Exploring columns of interest
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Columns to skip
columns_to_skip = ['Customer ID', 'City', 'Zip Code', 'Lat Long', 'Latitude', 'Longitude', 'Tenure Months', 'Monthly Charges', 'Total Charges', 'CLTV']

# Filter categorical columns to plot
categorical_cols = [col for col in df.columns if col not in columns_to_skip and df[col].dtype == 'object']

plots_per_row = 2
total_plots = len(categorical_cols)
rows = (total_plots + plots_per_row - 1) // plots_per_row

# Plotting bar graphs for each categorical column
for i in range(0, total_plots, plots_per_row):
    plt.figure(figsize=(15, 6))
    for j, col in enumerate(categorical_cols[i:i + plots_per_row]):
        plt.subplot(1, plots_per_row, j + 1)
        unique_counts = df[col].value_counts()
        sns.barplot(x=unique_counts.index, y=unique_counts.values)
        plt.title(f'Unique Value Counts for {col}')
        plt.xlabel('Unique Values')
        plt.ylabel('Count')
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
