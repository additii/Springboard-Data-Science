```python
import pandas as pd

# Load the data from CSV file
data = pd.read_csv('Starbucks_Nutrition (Starbucks_Calories)_Migrated Data.csv')

# Display the first few rows of the dataframe
print(data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting relevant columns for the pairplot
selected_columns = [
    'Calories', 
    'Sugar (g)', 
    'Total Carbohydrates (g)', 
    'Protein (g)', 
    'Total Fat (g)', 
    'Caffeine (mg)', 
    'BMI'
]

# Creating the pairplot
sns.pairplot(data[selected_columns])
plt.suptitle('Pairplot of Selected Starbucks Drink Features', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Selecting relevant columns for the correlation matrix
selected_columns = [
    'Calories', 
    'Sugar (g)', 
    'Total Carbohydrates (g)', 
    'Protein (g)', 
    'Total Fat (g)', 
    'Caffeine (mg)', 
    'BMI'
]

# Creating a subset of the data with the selected columns
selected_data = data[selected_columns]

# Calculating the correlation matrix
correlation_matrix = selected_data.corr()

# Creating the correlation matrix heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix Heatmap of Selected Starbucks Drink Features', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#1. Overview of Starbucks Drinksimport matplotlib.pyplot as plt
import seaborn as sns

# Number of unique Starbucks drinks
num_unique_drinks = data['Name'].nunique()

# Distribution of Starbucks drinks by serving size and caffeine content
plt.figure(figsize=(14, 6))
sns.scatterplot(x='Size in ounce', y='Caffeine (mg)', data=data, hue='Name', palette='viridis', s=100)
plt.title('Distribution of Starbucks Drinks by Serving Size and Caffeine Content')
plt.xlabel('Serving Size (ounce)')
plt.ylabel('Caffeine (mg)')
plt.legend(title='Starbucks Drink', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

print(f"Number of unique Starbucks drinks: {num_unique_drinks}")
```

```python
# 2. Nutritional Content of Starbucks Drinks
nutritional_columns = ['Calories', 'Sugar (g)', 'Total Carbohydrates (g)', 'Protein (g)', 'Total Fat (g)']

plt.figure(figsize=(16, 8))
sns.boxplot(data=data[nutritional_columns])
plt.title('Distribution of Nutritional Components in Starbucks Drinks')
plt.ylabel('Amount (g or kcal)')
plt.xticks(ticks=range(len(nutritional_columns)), labels=nutritional_columns, rotation=45)
plt.show()

# Correlation between calories and other nutritional components
correlation_matrix = data[nutritional_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Calories and Nutritional Components')
plt.show()
```

```python
#3. Exercise Intensity Required to Burn Calories

# Time to burn calories through different exercises
exercise_columns = ['Time to Burn - Bicycling', 'Time to Burn - Running', 'Time to Burn - Sitting', 'Time to Burn - Swimming']

plt.figure(figsize=(14, 8))
sns.boxplot(data=data[exercise_columns])
plt.title('Time Required to Burn Calories through Different Exercises')
plt.ylabel('Time (mins)')
plt.xticks(ticks=range(len(exercise_columns)), labels=['Bicycling', 'Running', 'Sitting', 'Swimming'])
plt.show()

# Relationship between caffeine content and time to burn calories
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Caffeine (mg)', y='Calories', data=data, hue='Time to Burn - Running', palette='viridis', s=100)
plt.title('Relationship between Caffeine Content, Calories, and Time to Burn Calories through Running')
plt.xlabel('Caffeine (mg)')
plt.ylabel('Calories')
plt.legend(title='Time to Burn (Running)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#4. Body Metrics and BMI# Distribution of body weight, height, and BMI
body_metrics = ['Weight', 'Height_m', 'BMI']

plt.figure(figsize=(16, 6))
sns.boxplot(data=data[body_metrics])
plt.title('Distribution of Body Metrics and BMI')
plt.ylabel('Value')
plt.xticks(ticks=range(len(body_metrics)), labels=body_metrics)
plt.show()

# Correlation between BMI and nutritional content
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='Calories', data=data, hue='Name', palette='viridis', s=100)
plt.title('Relationship between BMI, Calories, and Starbucks Drinks')
plt.xlabel('BMI')
plt.ylabel('Calories')
plt.legend(title='Starbucks Drink', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
