## Step 1: Data Wranggling<a id='1_Data_load_and_wrangling'></a>

### 1.1 Data Load and merging into one file<a id='1_1'></a>

In the data preperation step,

 1) I have loaded data and then checked for any nulls, duplicaes or missing values.
 2) Filtered Date time column for only 2019 and 2020 year.
     1) Converted 'Activate chat bot column to values 'Y' = 1 and 'N' = 0 so that I can check the correlation of it's values with the customer conversions from trial to premium in later stage.
 3)  Also, created new columns 
             'Monthly Percent Increase' of number of link clicks
             'Month-over-Month Growth' of number of link clicks
             'Usage start date' min(date time) pre account id
             'Usage end date' min(date time) pre account id
             'Usage age' Difference of Usage Start ate and Usage end date
 4) Exploring dataframe
 5) Change the grain of dataframe to account id by grouping all columns at that level in merged_data1.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load datasets
account_attributes = pd.read_excel("account_attributes.xlsx")
account_usage = pd.read_excel("account_usage.xlsx")

# Convert 'Date time' column to datetime data type
account_usage['Date time'] = pd.to_datetime(account_usage['Date time'])

# Merge datasets on 'Acct id'
merged_data = pd.merge(account_attributes, account_usage, on='Acct id')
```

### 1.2 Cleaning and filtering data<a id='1_2'></a>

```python
# Filter merged data for years 2019 and 2020
merged_data = merged_data[(merged_data['Date time'].dt.year.isin([2019, 2020]))]

# Convert 'Activate chat bot' column to numeric
merged_data['Activate chat bot'] = merged_data['Activate chat bot'].map({'Y': 1, 'N': 0})


# Handle missing values and creating new columns for later use.
merged_data.dropna(inplace=True)  # Drop rows with missing values
```

### 1.3 Adding new columns Monthly Percent Increase, Month-over-Month Growth, 'Usage start date', 'Usage end date', Usage age <a id='1_3'></a>

```python
# Adding new columns

#duplicating merged_data
merged_data_19_20 = merged_data

# Extract year and month from 'Date time' column
merged_data_19_20['Year'] = merged_data_19_20['Date time'].dt.year
merged_data_19_20['Month'] = merged_data_19_20['Date time'].dt.month

# Calculate total number of link clicks per account per month for each year
monthly_clicks_19_20 = merged_data_19_20.groupby(['Acct id', 'Year', 'Month'])['Number of link clicks'].sum().reset_index()
monthly_clicks_19_20.rename(columns={'Number of link clicks': 'Monthly Link Clicks'}, inplace=True)

# Calculate monthly percent increase compared to the previous month for each account
monthly_clicks_19_20['Monthly Percent Increase'] = monthly_clicks_19_20.groupby('Acct id')['Monthly Link Clicks'].pct_change() * 100

# Add the calculated percent increase as a new column to merged_data
# Merge only on 'Acct id' and 'Year' to avoid duplicating 'Date time' column
merged_data = pd.merge(merged_data, monthly_clicks_19_20[['Acct id', 'Year', 'Month', 'Monthly Percent Increase']], on=['Acct id', 'Year', 'Month'], how='left')

# Convert NaN to 0 in 'Monthly Percent Increase' column
merged_data['Monthly Percent Increase'] = merged_data['Monthly Percent Increase'].fillna(0)

# Convert 'Monthly Percent Increase' column to float with two decimal places
merged_data['Monthly Percent Increase'] = merged_data['Monthly Percent Increase'].astype(float).round(2)

# Calculate month-over-month growth compared to the previous month for each account
monthly_clicks_19_20['Month-over-Month Growth'] = monthly_clicks_19_20.groupby('Acct id')['Monthly Link Clicks'].diff()

# Add the calculated month-over-month growth as a new column to merged_data
# Merge only on 'Acct id' and 'Year' to avoid duplicating 'Date time' column
merged_data = pd.merge(merged_data, monthly_clicks_19_20[['Acct id', 'Year', 'Month', 'Month-over-Month Growth']], on=['Acct id', 'Year', 'Month'], how='left')

# Convert NaN to 0 in 'Month-over-Month Growth' column
merged_data['Month-over-Month Growth'] = merged_data['Month-over-Month Growth'].fillna(0)

# Find per customer usage start date and usage end date
usage_dates = merged_data.groupby('Acct id')['Date time'].agg(['min', 'max']).reset_index()
usage_dates.columns = ['Acct id', 'Usage start date', 'Usage end date']

# Calculate usage age
usage_dates['Usage age'] = (usage_dates['Usage end date'] - usage_dates['Usage start date']).dt.days

# Merge usage dates back to the merged data
merged_data = pd.merge(merged_data, usage_dates, on='Acct id', how='left')
```

### 1.5 Grouped all columns at Acct id level <a id='1_5'></a>

```python
# Copy merged_data to another DataFrame
df = merged_data.copy()

# Group by Acct id and aggregate the data
grouped_data = df.groupby('Acct id').agg({
    'Acct type': 'first',
    'Activate chat bot': 'first',
    'Converted to paid customer': 'first',
    'Date time': 'count',
    'Number of link clicks': 'sum',
    'Usage start date': 'first',
    'Usage end date': 'first',
    'Usage age': 'first'
}).reset_index()

# Rename the columns as required
grouped_data.columns = [
    'Acct id',
    'Acct type',
    'Activate chat bot',
    'Converted to paid customer',
    'Number of weeks active',
    'Total Number of link clicks',
    'Usage start date',
    'Usage end date',
    'Usage age'
]

# Calculate average number of link clicks per month per account
grouped_data['Average Number of link clicks per month'] = grouped_data['Total Number of link clicks'] / grouped_data['Number of weeks active']

# Assign the result back to merged_data
merged_data1 = grouped_data
```

### 2.2 Calculated Conversion Rate, Activation Rate, Engagement Rate and Average time to conversion And then divided these metric based on Acct Type<a id='2_2'></a>

```python
# Calculate the metrics
# 1. Conversion Rate
conversion_rate = (merged_data1['Converted to paid customer'].sum() / merged_data1['Converted to paid customer'].count()) * 100

# 2. Activation Rate
activation_rate = (merged_data1['Activate chat bot'].sum() / merged_data1['Activate chat bot'].count()) * 100

# 3. Engagement Rate
engagement_rate = merged_data1['Total Number of link clicks'].sum() / merged_data1['Acct id'].nunique()

# Filter merged data to exclude usage start date in January 2019 and usage end date in December 2020
filtered_data = merged_data1[~((merged_data1['Usage start date'].dt.year == 2019) & (merged_data1['Usage start date'].dt.month == 1)) & 
                            ~((merged_data1['Usage end date'].dt.year == 2020) & (merged_data1['Usage end date'].dt.month == 12))]

# Calculate Average Time to Conversion for the filtered data
average_time_to_conversion = (filtered_data['Usage end date'] - filtered_data['Usage start date']).dt.days.mean()

# Display the calculated metrics
print("Conversion Rate: {:.2f}%".format(conversion_rate))
print("Activation Rate: {:.2f}%".format(activation_rate))
#print("Engagement Rate: {:.2f}".format(engagement_rate))
print("Average Time to Conversion: {:.2f} days".format(average_time_to_conversion))
```

```python
# Group merged_data1 by account type
grouped_by_acct_type = merged_data1.groupby('Acct type')

# Calculate the metrics per account type
# 1. Conversion Rate per Account Type
conversion_rate_per_acct_type = (grouped_by_acct_type['Converted to paid customer'].sum() / grouped_by_acct_type['Converted to paid customer'].count()) * 100

# 2. Activation Rate per Account Type
activation_rate_per_acct_type = (grouped_by_acct_type['Activate chat bot'].sum() / grouped_by_acct_type['Activate chat bot'].count()) * 100

# 3. Engagement Rate per Account Type
engagement_rate_per_acct_type = grouped_by_acct_type['Total Number of link clicks'].sum() / grouped_by_acct_type['Acct id'].nunique()

# Filter merged_data1 to exclude usage start date in January 2019 and usage end date in December 2020
filtered_data_per_acct_type = merged_data1[~((merged_data1['Usage start date'].dt.year == 2019) & (merged_data1['Usage start date'].dt.month == 1)) & 
                            ~((merged_data1['Usage end date'].dt.year == 2020) & (merged_data1['Usage end date'].dt.month == 12))]

# Group filtered data per account type
filtered_grouped_by_acct_type = filtered_data_per_acct_type.groupby('Acct type')

# Calculate Average Time to Conversion per account type for the filtered data
average_time_to_conversion_per_acct_type = filtered_grouped_by_acct_type['Usage age'].mean()

# Display the calculated metrics per account type
print("Conversion Rate per Account Type:")
print(conversion_rate_per_acct_type)
print("\nActivation Rate per Account Type:")
print(activation_rate_per_acct_type)
#print("\nEngagement Rate per Account Type:")
#print(engagement_rate_per_acct_type)
print("\nAverage Time to Conversion per Account Type:")
print(average_time_to_conversion_per_acct_type)
```

## Step 4: Identifying Potential Customers Likely to Convert<a id='4_identify_potential_customers_likely_to_convert'></a>
Best model identified as RandomForestClassifier based on accuracy

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Define features for potential customer identification
features = ['Activate chat bot', 'Number of weeks active', 'Total Number of link clicks', 'Usage age', 'Acct type']

# Define preprocessor for numeric and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), ['Activate chat bot', 'Number of weeks active', 'Total Number of link clicks', 'Usage age']),
        ('cat', OneHotEncoder(), ['Acct type'])
    ])

# Initialize the pipeline with preprocessor and model
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', None)  # Model will be set dynamically
])

# List to store final potential customers across all simulations
all_final_potential_customers = []

# Repeat the simulation process 100 times
for simulation in range(100):
    potential_customers_list = []  # List to store potential customers for each account type
    
    # Simulate potential customer identification process for each account type
    for acct_type, data in [('ENT', ent_data), ('SMB', smb_data)]:
        X = data[features]
        y = data['Converted to paid customer']

        # Get the best model for the current account type
        best_model = models[best_models[acct_type]]

        # Set the model in the pipeline
        pipeline.set_params(model=best_model)

        # Train the pipeline on the current dataset
        pipeline.fit(X, y)

        # Predict the probability of conversion for each customer in the current dataset
        predicted_probabilities = pipeline.predict_proba(X)[:, 1]

        # Set a threshold probability (e.g., 0.5)
        threshold = 0.5

        # Identify potential customers likely to convert based on the threshold
        potential_customers = data[predicted_probabilities > threshold]

        # Remove duplicates from the list of potential customers
        potential_customers_unique = potential_customers.drop_duplicates(subset=['Acct id'])

        # Append potential customers to the list
        potential_customers_list.append(potential_customers_unique)

    # Concatenate potential customers from all account types for the current simulation
    final_potential_customers = pd.concat(potential_customers_list)
    
    # Append the final potential customers for the current simulation to the list
    all_final_potential_customers.append(final_potential_customers)

# Concatenate potential customers from all simulations
all_final_potential_customers = pd.concat(all_final_potential_customers)

# Save the final potential customers to a DataFrame
df_final_potential_customers = pd.DataFrame(all_final_potential_customers)

# Display the DataFrame
print("Final Potential Customers Likely to Convert (Unique) across all Simulations:")
print(df_final_potential_customers)
```

```python
from sklearn.metrics import confusion_matrix

# Define an empty confusion matrix
confusion_matrix_all = [[0, 0], [0, 0]]  # Format: [[true negatives, false positives], [false negatives, true positives]]

# Iterate through each simulation
for simulation in range(100):
    # Get the true labels for each account type
    y_true_ent = ent_data['Converted to paid customer']
    y_true_smb = smb_data['Converted to paid customer']

    # Get the predicted probabilities for each account type
    predicted_probabilities_ent = pipeline.predict_proba(ent_data[features])[:, 1]
    predicted_probabilities_smb = pipeline.predict_proba(smb_data[features])[:, 1]

    # Classify potential customers based on the threshold (e.g., 0.5)
    predicted_ent = (predicted_probabilities_ent > threshold).astype(int)
    predicted_smb = (predicted_probabilities_smb > threshold).astype(int)

    # Update the confusion matrix
    confusion_matrix_ent = confusion_matrix(y_true_ent, predicted_ent)
    confusion_matrix_smb = confusion_matrix(y_true_smb, predicted_smb)
    
    # Add current confusion matrix to the overall confusion matrix
    confusion_matrix_all += [[confusion_matrix_ent[0], confusion_matrix_ent[1]], [confusion_matrix_smb[0], confusion_matrix_smb[1]]]

# Average the confusion matrix over all simulations
confusion_matrix_all /= 100

# Display the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix_all)
```
