## Step 1: Data Preparation

In the data preperation step,

       1) I have checked for any nulls or missing values.
       2) Filtered Date time column for only 2019 and 2020 year.
       3) Converted 'Activate chat bot column to values 'Y' = 1 and 'N' = 0 so that I can check the correlation of it's values with the customer conversions from trial to premium in later stage.
       4)  Also, created two new columns 'Monthly Percent Increase' and 'Month-over-Month Growth' of number of link clicks.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load datasets
account_attributes = pd.read_excel("account_attributes.xlsx")
account_usage = pd.read_excel("account_usage.xlsx")

# Convert 'Date time' column to datetime data type
account_usage['Date time'] = pd.to_datetime(account_usage['Date time'])

# Merge datasets on 'Acct id'
merged_data = pd.merge(account_attributes, account_usage, on='Acct id')

# Filter merged data for years 2019 and 2020
merged_data = merged_data[(merged_data['Date time'].dt.year.isin([2019, 2020]))]

# Convert 'Activate chat bot' column to numeric
merged_data['Activate chat bot'] = merged_data['Activate chat bot'].map({'Y': 1, 'N': 0})


# Handle missing values and creating new columns for later use.
merged_data.dropna(inplace=True)  # Drop rows with missing values

# Adding new columns 'Monthly Percent Increase' and 'Month-over-Month Growth'
#----------------------------------------------------------------------------
merged_data_19_20 = merged_data

# Extract year and month from 'Date time' column
merged_data_19_20['Year'] = merged_data_19_20['Date time'].dt.year
merged_data_19_20['Month'] = merged_data_19_20['Date time'].dt.month

# Step 1: Calculate total number of link clicks per account per month for each year
monthly_clicks_19_20 = merged_data_19_20.groupby(['Acct id', 'Year', 'Month'])['Number of link clicks'].sum().reset_index()
monthly_clicks_19_20.rename(columns={'Number of link clicks': 'Monthly Link Clicks'}, inplace=True)

# Step 2: Calculate monthly percent increase compared to the previous month for each account
monthly_clicks_19_20['Monthly Percent Increase'] = monthly_clicks_19_20.groupby('Acct id')['Monthly Link Clicks'].pct_change() * 100

# Step 3: Add the calculated percent increase as a new column to merged_data
# Merge only on 'Acct id' and 'Year' to avoid duplicating 'Date time' column
merged_data = pd.merge(merged_data, monthly_clicks_19_20[['Acct id', 'Year', 'Month', 'Monthly Percent Increase']], on=['Acct id', 'Year', 'Month'], how='left')

# Convert NaN to 0 in 'Monthly Percent Increase' column
merged_data['Monthly Percent Increase'] = merged_data['Monthly Percent Increase'].fillna(0)

# Convert 'Monthly Percent Increase' column to float with two decimal places
merged_data['Monthly Percent Increase'] = merged_data['Monthly Percent Increase'].astype(float).round(2)

# Step 4: Calculate month-over-month growth compared to the previous month for each account
monthly_clicks_19_20['Month-over-Month Growth'] = monthly_clicks_19_20.groupby('Acct id')['Monthly Link Clicks'].diff()

# Step 5: Add the calculated month-over-month growth as a new column to merged_data
# Merge only on 'Acct id' and 'Year' to avoid duplicating 'Date time' column
merged_data = pd.merge(merged_data, monthly_clicks_19_20[['Acct id', 'Year', 'Month', 'Month-over-Month Growth']], on=['Acct id', 'Year', 'Month'], how='left')

# Convert NaN to 0 in 'Month-over-Month Growth' column
merged_data['Month-over-Month Growth'] = merged_data['Month-over-Month Growth'].fillna(0)

```

## Step 5: Building a Conversion Prediction Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns for one-hot encoding
categorical_columns = ['Acct type']  # 'Acct type' is a categorical column

# Check if all categorical columns are present in the dataset
missing_columns = [col for col in categorical_columns if col not in merged_data.columns]
if missing_columns:
    print("The following categorical columns are missing in the dataset:", missing_columns)
else:
    # One-hot encode categorical variables
    merged_data_encoded = pd.get_dummies(merged_data, columns=categorical_columns)

    # Features and target variable after encoding
    X_encoded = merged_data_encoded.drop(['Acct id', 'Date time', 'Activate chat bot'], axis=1)
    y_encoded = merged_data_encoded['Activate chat bot']

    # Split data into train and test sets
    X_train_encoded, X_test_encoded, y_train_encoded, y_test_encoded = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

    # Train and evaluate models with encoded data
    # Initialize models
    models = {
        'Logistic Regression': LogisticRegression(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier()
    }

    # Train and evaluate models
    results = {'Model': [], 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': []}
    for name, model in models.items():
        model.fit(X_train_encoded, y_train_encoded)
        y_pred = model.predict(X_test_encoded)
        accuracy = accuracy_score(y_test_encoded, y_pred)
        report = classification_report(y_test_encoded, y_pred, output_dict=True)
        results['Model'].append(name)
        results['Accuracy'].append(accuracy)
        results['Precision'].append(report['1']['precision'])
        results['Recall'].append(report['1']['recall'])
        results['F1-Score'].append(report['1']['f1-score'])

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    # Display results
    print(results_df)

    # Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(data=results_df, x='Model', y='Accuracy')
    plt.title('Model Comparison - Accuracy')
    plt.xlabel('Model')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=45)
    plt.show()
```

## Step 6: Identifying Potential Customers Likely to Convert

```python
# Choose the best performing model (e.g., Random Forest)
best_model = RandomForestClassifier()

# Train the best model on the entire dataset
best_model.fit(X_encoded, y_encoded)

# Predict the probability of conversion for each customer
predicted_probabilities = best_model.predict_proba(X_encoded)[:, 1]

# Set a threshold probability (e.g., 0.5)
threshold = 0.5

# Identify potential customers likely to convert based on the threshold
potential_customers = merged_data[predicted_probabilities > threshold]

# Remove duplicates from the list of potential customers
potential_customers_unique = potential_customers.drop_duplicates(subset=['Acct id'])

# Display potential customers without duplicates
print("Potential Customers Likely to Convert (Unique):")
print(potential_customers_unique)
```
